# Reservations and stays in the hotels of Lugano

## Auxiliars

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.tseries.offsets import *
```

```python
path = "../data/city_of_lugano/demographics/"
```

```python
def read_quarterly_data(file_name):
    dfs = []
    for year in np.arange(2008, 2019):
        for quarter in np.arange(1, 5):
            df = pd.read_json(path+"{0}/{1}/".format(year, quarter)+file_name+'.json')
            df['year'] = year
            df['quarter'] = quarter
            df['yq'] = pd.datetime(year,quarter*3-2,1)
            dfs.append(df)

    df = pd.concat(dfs)
    return df
```

## Analysis

### Pre-processing

#### Translating columns and parsing date

```python
df = pd.read_csv("../data/h_benchmark/Lugano-prenotazioniv2.csv", sep=';', dtype=str)

df.columns = [
    "id",
    "date-arrived",
    "date-left",
    "duration",
    "date-creation",
    "date-cancellation",
    "num-reserved-rooms",
    "num-guests",
    "overall-revenue",
    "rooms-revenue",
    "room-type",
    "state",
    "treatment",
    "channel",
    "convention",
    "nationality",
]

tra = [
    ("gen","1"), ("feb","2"), ("mar","3"), ("apr","4"), ("mag","5"), ("giu","6"),
    ("lug","7"), ("ago","8"), ("set","9"), ("ott","10"), ("nov","11"), ("dic","12"),
]

for dates in ['date-arrived', 'date-left']:
    for tr in tra:
        df[dates] = df[dates].apply(lambda x : x.replace(*tr))
    df[dates] = pd.to_datetime(df[dates], format='%d-%m-%y')

df.head(5)
```

```python
print(df.shape)
print(df.drop_duplicates().shape)
print(df.dropna().shape)
```

#### Taking out useless states (cancelled and no show)

```python
dfattended = df[df['state'] != "Cancellata"]
dfattended = dfattended[dfattended['state'] != "No show"]
dfattended.head()
```

#### Function to calculate frequency for a specific limit

```python
def graph_frequency(df, label, drop=None, limit=None, rangePlot=None):
    dfreturn = df[label].value_counts()
    if limit != None:
        dfreturn = dfreturn[dfreturn.index.astype(int) < limit]
        
    if drop != None:
        for d in drop:
            dfreturn.drop(labels=d)

    if rangePlot != None:
        axreturn = dfreturn[rangePlot[0]:rangePlot[1]].plot.bar(figsize=(15,10), fontsize=13)
    else:
        axreturn = dfreturn.plot.bar(figsize=(15,10), fontsize=13)
        
    axreturn.set_title("Frequency of %s" % label, fontdict={'fontsize': 16, 'fontweight': 'bold'})
    return axreturn
```

### Relating date of arrival with booking purpose and nationality

```python
mask = dfattended['date-arrived'] >= pd.datetime(2019, 5, 1)
mask = mask & (dfattended['date-arrived'] <= pd.datetime(2019, 7, 31))
df2 = dfattended[mask].copy()
df2 = df2[df2['date-arrived'] <= pd.datetime(2019, 7, 31)]
```

#### Convention frequency

```python
df2.groupby(['date-arrived', 'convention']).size().unstack().rolling(7,2).mean().\
plot(kind='line', figsize=(15,10))
plt.grid(True)
```

#### Nationality frequency

```python
df2.groupby(['date-arrived', 'nationality']).size().unstack().rolling(7,2).mean().\
plot(kind='line', figsize=(15,10))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True)
```
